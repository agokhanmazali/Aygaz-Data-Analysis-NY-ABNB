```python
correlation_matrix = df_new.corr()

# Plot the correlation matrix as a heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Heatmap of Airbnb NYC 2019 Data')
plt.show()
```

* As shown on the correlation graph, there is no meaningful correlations because there is no value greater than 0.5 or less than -0.5

```python
plt.figure(figsize=(10,6))
sns.histplot(data=df_new, x='price', bins=100)
plt.title('Price Distribution of Airbnb Listings')
plt.xlabel('Price')
plt.ylabel('Count')
plt.xlim([0,1000])  # Limiting to 1000 for better visualization
plt.show()
```

* The distribution is heavily skewed to the right, indicating that most listings are in the lower price range.

* There is a large number of listings priced around $100-$200, which could be considered the "typical" price range for an Airbnb in this dataset.

* There are fewer listings in the higher price range (above $200).

* The plot is limited to a price of $1000 for better visualization, so listings above this price are not shown.

```python
plt.figure(figsize=(10,8))
sns.scatterplot(data=df_new, x='longitude', y='latitude', hue='neighbourhood_group', palette='bright')
plt.title('Geographical Distribution of Airbnb Listings in NYC')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.legend(title='Neighbourhood Group')
plt.show()
```

* Listings are concentrated in Manhattan, Brooklyn, and parts of Queens, which aligns with these areas being popular tourist destinations.
* Manhattan listings are concentrated in the central part of the island, where many of the city's attractions are located.
* Brooklyn listings are spread throughout the borough, with a higher concentration in the north, closer to Manhattan.
* Queens has fewer listings, and they are mainly located close to Manhattan and Brooklyn.
* Staten Island and the Bronx have relatively fewer listings.

```python
# Plotting the count of each room type
plt.figure(figsize=(8,6))
sns.countplot(data=df_new, x='room_type', palette='bright')
plt.title('Count of Each Room Type')
plt.xlabel('Room Type')
plt.ylabel('Count')
plt.show()
```

* "Entire home/apt" is the most common type of listing. This suggests that most hosts prefer to rent out their entire property rather than sharing it.
* "Private room" is the second most common type of listing. These listings typically involve guests having their own room but sharing some spaces, such as the kitchen or bathroom, with others.
* "Shared room" is the least common type of listing. These are listings where guests share a room, typically a living room or bedroom, with others.

```python
# Calculate average price for each room type
average_price_by_room_type = df_new.groupby('room_type')['price'].mean()

# Plot average price for each room type
plt.figure(figsize=(8,6))
sns.barplot(x=average_price_by_room_type.index, y=average_price_by_room_type.values, palette='bright')
plt.title('Average Price for Each Room Type')
plt.xlabel('Room Type')
plt.ylabel('Average Price')
plt.show()
```

*

* "Entire home/apt" has the highest average price among the room types. This makes sense as guests renting an entire home or apartment would expect to pay more than for a single room.

* "Private room" has the second highest average price. In this type of listing, guests have their own private room but share some spaces, such as the kitchen or bathroom, with others.

* "Shared room" has the lowest average price. These are listings where guests share a room, typically a living room or bedroom, with others.

```python
# Define a color palette with different shades of blue
blue_palette = sns.color_palette("Blues_r", df['room_type'].nunique())

plt.figure(figsize=(10, 6))
sns.barplot(x='neighbourhood_group', y='price', hue='room_type', data=df_new, errorbar=None, palette=blue_palette)
plt.title('Average Price of Room Types in Neighbourhood Groups')
plt.xlabel('Neighbourhood Group')
plt.ylabel('Average Price')
plt.xticks(rotation=45)
plt.show()

```

From the plot, we can observe the following:

* Across all neighbourhood groups, "Entire home/apt" is the most expensive room type on average, followed by "Private room", and then "Shared room".

* Manhattan has the highest average price for all room types among the neighbourhood groups.

* Brooklyn and Queens have similar average prices for private rooms and shared rooms, but Brooklyn has a higher average price for entire homes/apartments.

* Staten Island and the Bronx generally have lower average prices compared to the other neighbourhood groups.

```python
# Group by neighbourhood_group and calculate the mean of price and availability_365
grouped = df_new.groupby('neighbourhood_group').agg({'price':'mean','availability_365':'mean'})

# Create the bar plot
ax = grouped.plot.bar(alpha=0.7, figsize=(10, 6), color=['skyblue', 'blue'])
ax.set_title("Mean Price and Availability by Neighbourhood Group")
ax.set_xlabel("Neighbourhood Group")
ax.set_ylabel("Mean Value")
plt.legend(['Price', 'Availability'])
plt.xticks(rotation=45)
plt.show()
```

* Across all neighbourhood groups, "Entire home/apt" is the most expensive room type on average, followed by "Private room", and then "Shared room".
* Manhattan has the highest average price for all room types among the neighbourhood groups.
* Brooklyn and Queens have similar average prices for private rooms and shared rooms, but Brooklyn has a higher average price for entire homes/apartments.
* Staten Island and the Bronx generally have lower average prices compared to the other neighbourhood groups.

```python
# Calculate the number of hosts for each number of listings they have
host_listings_counts = df_new['calculated_host_listings_count'].value_counts().sort_index()

# Plot the number of hosts for each number of listings
plt.figure(figsize=(12,6))
sns.barplot(x=host_listings_counts.index, y=host_listings_counts.values, color='b')
plt.title('Number of Listings per Host')
plt.xlabel('Number of Listings')
plt.ylabel('Number of Hosts')
plt.xscale('log')  # Use a logarithmic scale for better visualization
plt.show()

# Calculate the number and percentage of 'superhosts' (hosts with more than 1 listing)
num_superhosts = (df_new['calculated_host_listings_count'] > 1).sum()
perc_superhosts = num_superhosts / len(df_new) * 100

num_superhosts, perc_superhosts
```

* The bar plot above shows the number of hosts (y-axis) for each number of listings they have (x-axis). The x-axis is in a logarithmic scale for better visualization since there are hosts with a large number of listings.

* Here are some insights from the plot and the calculations:

* Most hosts have only 1 listing. This is indicated by the tall bar at the left of the plot.
* As the number of listings per host increases, the number of hosts with that many listings decreases. This is a common characteristic in many real-world distributions and is often referred to as a long-tail distribution.
* There are some hosts with a very large number of listings, but they are relatively rare.
* From the calculations:

* There are 16,592 'superhosts', which are defined here as hosts with more than 1 listing.
* These 'superhosts' make up approximately 33.93% of all hosts.
* These insights give us an understanding of how listings are distributed among hosts. A majority of hosts have only one listing, but a significant proportion (about one-third) have multiple listings.

```python
average_price_by_neighbourhood_group = df_new.groupby('neighbourhood_group')['price'].mean().sort_values(ascending=False)

# Plot average price for each neighbourhood_group
plt.figure(figsize=(8,6))
sns.barplot(x=average_price_by_neighbourhood_group.index, y=average_price_by_neighbourhood_group.values, palette='bright')
plt.title('Average Price for Each Neighbourhood Group')
plt.xlabel('Neighbourhood Group')
plt.ylabel('Average Price')
plt.show()
```

* Manhattan has the highest average price among all the neighbourhood groups. This makes sense considering that Manhattan is often the most expensive area in New York City in terms of real estate.

* Brooklyn has the second highest average price, followed closely by Staten Island. Brooklyn has been becoming increasingly popular and expensive, which might explain its position.

* The Bronx and Queens have the lowest average prices. These areas are generally less expensive than Manhattan and Brooklyn.
